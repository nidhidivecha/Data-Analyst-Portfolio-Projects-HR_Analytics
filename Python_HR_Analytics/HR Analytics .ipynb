# DATA PRE-PROCESSING

```python
df.isnull().sum()
```

## Categorical Data Handling

```python
df["BusinessTravel"].replace("Travel_Rarely", 1, inplace = True)
df["BusinessTravel"].replace("Travel_Frequently", 2, inplace = True)
df["BusinessTravel"].replace("Non-Travel", 3, inplace = True)
```

```python
df.head()
```

```python
df["OverTime"]
```

```python
df["OverTime"].replace("Yes", 1, inplace = True)
df["OverTime"].replace("No", 2, inplace = True)
```

```python
df["MaritalStatus"]= df["MaritalStatus"].map({"Yes": 1, "No" :0})
```

```python
df["Gender"]= df["Gender"].map({"Male": 1, "Female" :2})
```

```python
df["JobRole"]= df["JobRole"].map({"Laboratory Technician": 8, "Sales Executive" :7, "Research Scientist": 6, "Sales Representive" :5, "Human Resources": 4, "Manufacturing Director" :3, "Healthcare Representative": 2, "Manager" :1, "Research Director" :0})
```

```python
df.info()
```

## Feature Selection

### Dropping unique ones

```python
df.drop(["EmployeeCount", "EmployeeNumber", "Over18", "StandardHours", "MaritalStatus"], axis = 1, inplace = True)
df.dropna(inplace = True)
```

## Model Creation

```python
x = df.drop(["Attrition"],axis = 1)
y = df["Attrition"]
```

## How to treate imbalanced datasets

Oversampling - SMOTE 

Undersampling - RandomUnderSampler

```python
# !pip install -U imbalanced-learn
```

```python
from imblearn import under_sampling, over_sampling
from imblearn.over_sampling import SMOTE
```

```python
x.info()
```

```python
from collections import Counter
from imblearn.over_sampling import SMOTE
sm = SMOTE()
print("unbalanced data:", Counter(y))
x_sm, y_sm = sm.fit_resample(x,y)
print("balanced data:   :", Counter(y_sm))
```

## Train-Test-Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x_sm, y_sm, random_state = 42, test_size=0.25) 
```

## Model Building

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier() 
dt.fit(x_train, y_train)
```

## Prediction

```python
y_pred=dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score
```

```python
print(classification_report(y_test,y_pred))
```

## checking data is fully trained or not

```python
train_pred=dt.predict(x_train)
```

```python
print(classification_report(y_train,train_pred))
```

## Random Forest Technique

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=100)
```

```python
rfc.fit(x_train,y_train)
```

```python
y_prediction=rfc.predict(x_test)
```
